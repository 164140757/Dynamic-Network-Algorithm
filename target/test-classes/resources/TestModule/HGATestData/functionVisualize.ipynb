```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
from cycler import cycler
```

```python
import math
def simFunction(x):
    return 1./(1.-1./np.log(x))
# 1 and 0 is not allowed 
x_1 = np.arange(0.001, math.e, 0.01)
x_2 = np.arange(math.e+0.1, 30., 0.01)

ax = plt.axes()
ax.set_facecolor('white')

plt.grid(False)
plt.plot(x_1,simFunction(x_1), 'g-')
plt.plot(x_2,simFunction(x_2),'b-')

plt.xlim((0,20))
plt.ylim((-20,20))
plt.axvline(x=math.e, ymin=-300, ymax=300,c='r', linewidth=0.5)    
plt.axhline(y=1,c='r',ls='--', linewidth=1)    
plt.axhline(y=-1,c='r',ls='--', linewidth=1)    
plt.show()
```
